```python
%matplotlib inline
import matplotlib.pyplot as plt
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis,color="magenta", label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
# y axis months going from dec to jan need to do right order
plt.gca().invert_yaxis()
```

```python
#bar charts using Object-Oriented Approach
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis,color="cyan", linewidth=2,label='New York')
ax.invert_yaxis()
ax.legend()
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis,color="cyan", linewidth=2,label='Chicago')
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
ax.invert_yaxis()
```
