```python
import matplotlib.pyplot as plt
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot( y_axis, x_axis,'o',color="red", label='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Months")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.legend()
# plt.scatter(x_axis, y_axis,'o')

```

```python
# x axis,y axis , size , color 4 factors
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#list comprehension i is variable represents items in y_axis list[]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,color='royalblue',label='Boston')
# ax.scatter(y_axis,x_axis,)
# ax.scatter(y_axis,x_axis,edgecolor='k',to make transparent alpha=0.20)
ax.scatter(y_axis,x_axis,color='pink',edgecolor='k', s=[i * 5 for i in y_axis])
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Months")
ax.set_xlim(0,50)
ax.legend()
```
