```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot matlab
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan","June"") or plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# to change pixel size use plt.figure(figsize=(w,h)) or plt.subplots(figsize=(20, 10))
# Add the legend.
plt.legend()
```

```python
#same plot with Matplotlib Object-Oriented Functions 
# to get a chart figure .
fig, ax=plt.subplots(figsize=(10,5))
ax.plot(x_axis, y_axis,marker="D", color="green", linewidth=2,label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
#ax.set_xlim(Jan, June)
#plt.xlim("Jan","June"") or plt.xlim(0,5)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# to change pixel size use plt.figure(figsize=(w,h)) or plt.subplots(figsize=(20, 10))

# Add the legend.
ax.legend()
```
